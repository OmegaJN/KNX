```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

# Загрузка данных
df = pd.read_csv("data2.csv", sep=",")
x = df.iloc[:, 7].values

# Задание числа предыдущих значений
n_lags = 5

# Создание матрицы предыдущих значений
X = np.zeros((len(x) - n_lags, n_lags))
for i in range(n_lags, len(x)):
    X[i-n_lags] = x[i-n_lags:i]

# Создание вектора выходных значений
y = np.sign(x[n_lags:])

# Обучение модели
clf = LogisticRegression()
clf.fit(X, y)

# Предсказание вероятностей для последних n_lags значений
last_n_vals = x[-n_lags:].reshape(1, -1)
probabilities = clf.predict_proba(last_n_vals)
print('Вероятность появления значения -1:', probabilities[0,0])
print('Вероятность появления значения 1:', probabilities[0,1])
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv("data2.csv", sep=",")
x = df.iloc[:, 7].values

# Задание числа предыдущих значений
n_lags = 10

# Создание матрицы предыдущих значений
X = np.zeros((len(x) - n_lags, n_lags))
for i in range(n_lags, len(x)):
    X[i-n_lags] = x[i-n_lags:i]

# Создание вектора выходных значений
y = np.sign(x[n_lags:])


# Обучение модели
clf = LogisticRegression()
clf.fit(X, y)

# Предсказание вероятностей для значений от 1 до 50
for i in range(1, 51):
    vals = x[-n_lags-i:-i].reshape(1, -1)
    probabilities = clf.predict_proba(vals)
    print('Для последних {0} значений:'.format(i))
    print('Вероятность появления значения -1:', probabilities[0,0])
    print('Вероятность появления значения 1:', probabilities[0,1])

# Создание списков значений вероятностей
prob_1 = []
prob_minus_1 = []
for i in range(1, 51):
    vals = x[-n_lags-i:-i].reshape(1, -1)
    probabilities = clf.predict_proba(vals)
    prob_minus_1.append(probabilities[0,0])
    prob_1.append(probabilities[0,1])

# Построение столбикового графика вероятностей
plt.bar(range(1, 51), prob_1, color='blue', label='Вероятность 1')
plt.bar(range(1, 51), prob_minus_1, color='red', label='Вероятность -1')
plt.xlabel('Количество предыдущих значений')
plt.ylabel('Вероятность')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv("data2.csv", sep=",")
x = df.iloc[:, 7].values

# Задание числа предыдущих значений
n_lags = 10

# Создание матрицы предыдущих значений
X = np.zeros((len(x) - n_lags, n_lags))
for i in range(n_lags, len(x)):
    X[i-n_lags] = x[i-n_lags:i]

# Создание вектора выходных значений
y = x[n_lags:]


# Обучение модели
clf = RandomForestClassifier()
clf.fit(X, y)

# Предсказание вероятностей для значений от 1 до 50
for i in range(1, 51):
    vals = x[-n_lags-i:-i].reshape(1, -1)
    probabilities = clf.predict_proba(vals)
    print('Для последних {0} значений:'.format(i))
    print('Вероятность появления значения -1:', probabilities[0,0]/(probabilities[0,0]+probabilities[0,1]))
    print('Вероятность появления значения 1:', probabilities[0,1]/(probabilities[0,0]+probabilities[0,1]))
    
# Создание списков значеинй вероятностей
prob_1 = []
prob_minus_1 = []
for i in range(1, 51):
    vals = x[-n_lags-i:-i].reshape(1, -1)
    probabilities = clf.predict_proba(vals)
    prob_minus_1.append(probabilities[0,0]/(probabilities[0,0]+probabilities[0,1]))
    prob_1.append(probabilities[0,1]/(probabilities[0,0]+probabilities[0,1]))

# Построение столбикового графика вероятностей
plt.bar(range(1, 51), prob_1, color='blue', label='Вероятность 1')
plt.bar(range(1, 51), prob_minus_1, color='red', label='Вероятность -1')
plt.xlabel('Количество предыдущих значений')
plt.ylabel('Вероятность')
plt.legend()
plt.show()
```

Пример кода с калибровкой модели:

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv("data2.csv", sep=",")
x = df.iloc[:, 7].values

# Задание числа предыдущих значений
n_lags = 10

# Создание матрицы предыдущих значений
X = np.zeros((len(x) - n_lags, n_lags))
for i in range(n_lags, len(x)):
    X[i-n_lags] = x[i-n_lags:i]

# Создание вектора выходных значений
y = np.sign(x[n_lags:])
y_new = np.where(y == -1, 0, 1)
# Обучение модели
clf = LogisticRegression()
clf.fit(X, y)

# Предсказание вероятностей для значений от 1 до 50
for i in range(1, 51):
    vals = x[-n_lags-i:-i].reshape(1, -1)
    probabilities = clf.predict_proba(vals)
    print('Для последних {0} значений:'.format(i))
    print('Вероятность появления значения -1:', probabilities[0,0])
    print('Вероятность появления значения 1:', probabilities[0,1])

# Получение калибровочных кривых
# prob_true, prob_pred = calibration_curve(y, clf.predict_proba(X)[:,1], n_bins=5, strategy='quantile')
y_means, proba_means = calibration_curve(y_new, clf.predict_proba(X)[:,1], n_bins=5, strategy='quantile')
# Построение графика калибровочных кривых
plt.plot(y_means, proba_means, marker='.', label='Логистическая регрессия')
plt.plot([0, 1], [0, 1], linestyle='--', label='Идеальная калибровка')
plt.xlabel('Средняя предсказанная вероятность')
plt.ylabel('Фактическая вероятность')
plt.legend()
plt.show()
```

Нейронная сеть:

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv("data2.csv", sep=",")
x = df.iloc[:, 7].values

# Задание числа предыдущих значений
n_lags = 10

# Создание матрицы предыдущих значений
X = np.zeros((len(x) - n_lags, n_lags))
for i in range(n_lags, len(x)):
    X[i-n_lags] = x[i-n_lags:i]

# Создание вектора выходных значений
y = np.sign(x[n_lags:])
y_new = np.where(y == -1, 0, 1)

# Обучение модели
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=n_lags))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X, y_new, epochs=100, batch_size=64)

# Предсказание вероятностей для значений от 1 до 50
for i in range(1, 51):
    vals = x[-n_lags-i:-i].reshape(1, -1)
    probabilities = model.predict(vals)
    print('Для последних {0} значений:'.format(i))
    print('Вероятность появления значения -1:', 1 - probabilities[0,0])
    print('Вероятность появления значения 1:', probabilities[0,0])

# Получение калибровочных кривых
y_means, proba_means = calibration_curve(y_new, model.predict(X).flatten(), n_bins=5, strategy='quantile')
# Построение графика калибровочных кривых
plt.plot(y_means, proba_means, marker='.', label='Нейронная сеть')
plt.plot([0, 1], [0, 1], linestyle='--', label='Идеальная калибровка')
plt.xlabel('Средняя предсказанная вероятность')
plt.ylabel('Фактическая вероятность')
plt.legend()
plt.show()
```
